## District Data

```python
import pandas as pd
import numpy as np
import math
```

```python
District = pd.read_csv('AllUSDistricts.CSV')
AllDistricts = District[["state_code","district"]]
AllDistricts = AllDistricts.rename(columns={'state_code':'State','district':'District'})
AllDistricts["State-District"] = AllDistricts["State"].map(str) + "-" + (AllDistricts["District"]).map(str)
AllDistricts.head()
```

```python
House = pd.read_csv("NASAHouseCommittees.csv")
House = House.drop(["Name"],axis=1)
House["HouseCommitties"] = House["House Science, Space and Technology Committee"] + House["House Space and Aeronautics Subcommittee"] + House["House CJS"]
House
```

```python
Centers = pd.read_csv('NASACenters_District.csv')
CentName = pd.read_csv('NASACenters_District.csv')
Centers = Centers.drop(["Zip"], axis=1)
Centers["Center"] = 1 
Centers = Centers.groupby(["State","District"]).sum()
Centers
```

```python
ConHouse = pd.read_csv("District_Cleaned_Nasa_Data.csv")
ConHouse19 = ConHouse[ConHouse.Year == 2019].reset_index()
ConHouse19 = ConHouse19.drop(["Year","index"],axis=1)
ConHouse19
```

```python
ACleanDistricts = pd.merge(AllDistricts,Centers, on=["State","District"],how='left')
BCleanDistricts = pd.merge(ACleanDistricts,House, on=["State","District"],how='left')
CCleanDistricts = pd.merge(BCleanDistricts,ConHouse19, on=["State","District"],how='left')
CleanDistricts = pd.merge(CCleanDistricts,CentName, on=["State","District"],how='left')
CleanDistricts = CleanDistricts.fillna(0)
CleanDistricts["NatLog_Award"] = np.where(CleanDistricts['Award']==0, 0.0, np.log(CleanDistricts.Award))
CleanDistricts["Log10_Award"] = np.where(CleanDistricts['Award']==0, 0.0, np.log10(CleanDistricts.Award))
CleanDistricts.Center_Abbreviation.replace(0, np.nan, inplace=True)
CleanDistricts.Center_Name.replace(0, np.nan, inplace=True)
CleanDistricts
```

## State Data

```python
States = pd.read_csv('StateList.csv')
States = States.drop(["State","Abbrev"],axis=1)
States = States.rename({"Code":"State"},axis=1)
States
```

```python
StateCenters = Centers
StateCenters = StateCenters.groupby(["State"]).sum()
StateCenters
```

```python
Senate = pd.read_csv("NASASenateCommittees.csv")
Senate = Senate.drop(["Name"],axis=1)
Senate.sort_values(by=["State"],inplace=True)
Senate["SenateCommittess"] = Senate["Senate Commerce, Science, and Transportation"] + Senate["Senate Aviation and Space Subcommittee"] + Senate["Senate CJS"]
Senate
```

```python
ConSenate = pd.read_csv("State_Cleaned_Nasa_Data.csv")
ConSenate19 = ConSenate[ConSenate.Year == 2019].reset_index()
ConSenate19 = ConSenate19.drop(["Year","index"],axis=1)
ConSenate19
```

```python
ACleanState = pd.merge(States,StateCenters, on=["State"],how='left')
BCleanState = pd.merge(ACleanState,ConSenate19, on=["State"],how='left')
CleanStates = pd.merge(BCleanState,Senate, on=["State"],how='left')
CleanStates = CleanStates.fillna(0)
CleanStates["NatLog_Award"] = np.where(CleanStates['Award']==0, 0.0, np.log(CleanStates.Award))
CleanStates["Log10_Award"] = np.where(CleanStates['Award']==0, 0.0, np.log10(CleanStates.Award))
CleanStates
```

```python
CleanStates.to_csv(r'Cleaned_States19.csv')
CleanDistricts.to_csv(r'Cleaned_Districts19.csv')
```
