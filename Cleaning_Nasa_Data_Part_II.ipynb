```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Cleaned_Nasa_Data.csv')

## Remove quotes around district number
data["District"] = data["District"].str.replace('\'', '')
data['Center'] = data['Center'].str.split().str[0].str.strip()
data = data.drop(['Contractor'], axis = 1)
data.head()
```

```python
data = data[['Center', 'State', 'District',  'State-District', 'Year','DSY','Award']]
data.head()
```

```python
Statelist = ['NY', 'VA', 'TX', 'CA', 'WI', 'MN', 'NC', 'AZ', 'CT', 'FL', 'MI',
       'OH', 'MD', 'AL', 'NJ', 'IL', 'MS', 'OR', 'RI', 'NV', 'MA', 'DC',
       'CO', 'NH', 'DE', 'TN', 'PA', 'WA', 'NM', 'LA', 'KS', 'OK', 'MT',
       'IN', 'GA', 'AK', 'UT', 'IA', 'SC', 'ID', 'AR', 'VT', 'WV', 'MO',
       'WY', 'ME', 'SD', 'HI', 'ND', 'KY', 'NE', 'PR', 'UK', 'VI', 'GU',
       'AS']
Truestate = pd.read_csv("StateList.csv")
Truestate = list(Truestate["Code"])

NotState = []

for item in Statelist:
    if item not in Truestate:
        NotState.append(item)

NotState
#len(Truestate)
```

```python

AtLarge = ["AK","DE","MT","ND","SD","VT","WY"]
```

```python
for terr in NotState:
    #print(len(data))
    data = data[data.State != terr]
    #print(len(data))
```

```python
States = data.drop(["District","State-District","DSY"], axis = 1)
StateYear = States.groupby(["Year", "State"]).sum()
```

```python
StateYear.to_csv(r'State_Cleaned_Nasa_Data.csv', index=True)
```

```python
Districts = data.drop(data[(~data.State.isin(AtLarge)) & (data.District == "00")].index)
Districts
```

```python
DistrictGroup = Districts.groupby(["Year","State","District"]).sum()
DistrictGroup.to_csv(r'District_Cleaned_Nasa_Data.csv', index=True)
```
